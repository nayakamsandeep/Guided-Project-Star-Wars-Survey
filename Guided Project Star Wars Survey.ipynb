```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1',
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
})
star_wars.head()
```

```python
means = star_wars[['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4','ranking_5', 'ranking_6']].mean()

print(means)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.bar(range(1,7), means, width = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

seen_sum = star_wars[star_wars.columns[3:9]].sum()

fig, ax = plt.subplots()

ax.bar(range(1,7), seen_sum, width = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

means_male = males[['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4','ranking_5', 'ranking_6']].mean()
means_female = females[['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4','ranking_5', 'ranking_6']].mean()

fig, (ax1, ax2) = plt.subplots(2, figsize = (5, 10))

ax1.bar(range(1,7), means_male, width = 0.5)
ax1.set_title('Male Rankings')
ax2.bar(range(1,7), means_female, width = 0.5)
ax2.set_title('Female Rankings')
plt.show()
```

```python
import matplotlib.pyplot as plt
males_seen_sum = males[males.columns[3:9]].sum()
females_seen_sum = females[females.columns[3:9]].sum()

fig, (ax1,ax2) = plt.subplots(2, figsize = (5, 10))

ax1.bar(range(1,7), males_seen_sum, width = 0.5)
ax1.set_title('Most Seen Movies By Males')
ax2.bar(range(1,7), females_seen_sum, width = 0.5)
ax2.set_title('Most Seen Movies By Females')
plt.show()
```
